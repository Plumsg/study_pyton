### 딕셔너리


```python
# 딕셔너리명 = {key:value, key:value, key:value ... }
# key는 변하지 않는 값을 사용
# value는 변하는 값, 변하지 않는 값 모두 사용 가능
```

```python
a = {}
b = {"name" :"KimMihee", "pn": "010-1111-2222", "age" : 20}
c = {1:"abcv", 2:456}
print(type(a))
print(b)
print(c)
```

```python
# 딕셔너리 값 추가하기
# 딕셔너리명[key] = value
dic1 =  {"name" :"KimMihee", "pn": "010-1111-2222", "age" : 20}
dic1['birth'] = '0705'
dic1['pn'] = '010-3333-4444'
print(dic1)
```

```python
# 딕셔너리 값 가져오기 1
print(dic1['name'])

# 딕셔너리 값 가져오기 2
# 딕셔너리명.get(key)
print(dic1.get('name'))
```

```python
# print(dic1['email'])
print(dic1.get('email'))
```

```python
if dic1.get('email') == None :
    print('없는 데이터')
```

```python
# 딕셔너리 key만 가져오기
# 딕셔너리명.keys()
keyLi=list(dic1.keys())
for k in dic1.keys():
    print(k)
```

```python
# 딕셔너리 value만 가져오기
# 딕셔너리명.values()
valueLi = list(dic1.values())
print(valueLi[0])
for value in dic1.values() :
    print(value)
```

```python
dic1.items()
```

```python
for key, value in dic1.items() :
    print(key, value)
```

```python
# 딕셔너리와 in 키워드 사용
print('email' in dic1)
print('name' in dic1)
```

```python
# 딕서녀리 값 삭제
# del 딕셔너리명[key]
del dic1['age']
```
