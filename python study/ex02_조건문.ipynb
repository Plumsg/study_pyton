### 조건문

#### if 문
- if 조건식 :
- (들여쓰기)실행문장
- 주의사항
    - 조건식에는 논리형 결과를 써줘야한다 (True,False)
    - 반드시 콜론
    - 반드시 들여쓰기 -> 조건식의 실행범위 영역을 표시

```python
if True : 
    print("실행문장실행")
```

```python
# 조건식이 False 였을때는 아무 실행도 하지 않고 끝난다.
if False :
    print("실행문장실행")
```

- 들여쓰기의 중요성!

```python
if True:
    print("조건식이 참이었을때 실행문장")
print("if문 바깥쪽에 있는 문장")
```

```python
if False :
    print("조건식이 거짓이었을때 실행문장")
print("if문 바깥쪽에 있는 문장")
# 들여쓰기가 되어있지 않으면 if 의 영향을 받지 않는다
```

```python
money = 11000
if money >10000 :
    print("택시를 탄다")
```

#### if else 문
- if 조건식: 
- (들여쓰기)실행문장1
- else:
- (들여쓰기)실행문장2
- 주의사항
    - if, else는 같은 선상에 있어야한다
    - 콜론, 들여쓰기
    - else는 참이 아니였을 모든 경우를 포함하기 때문에 조건식을 적지 않는다

```python
# 단순 if 문과의 차이점: 조건식이 False 였을경우 실행문장의 유무차이
if False:
    print("실행문장1")
else:
    print("실행문장2")
```

```python
money = 11000
if money >= 10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

```python
num = int(input())

if num % 3 == 0 and num % 5== 0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다")
```

```python
num = int(input("사려는 상품의 갯수를 입력하세요"))
#11개 미만일때
if num<11:
    price = 10000*num
else:
    price = 9000*num
print(f"가격은 {price}원 입니다")
```

### if elif else 문
- 여러개의 조건을 사용할 때

- 주의
    - 반드시 else문을 작성해야한다 xxxxx

```python
# 돈이 10000원 이상일 경우 택시를 탄다
money = int(input())

if money >=10000 :
    print("택시를 탄다")
    
# 돈이 10000원 미만 1000원 이상일 경우 버스를 탄다
elif 1000<= money<= 10000:
    print("걸어간다")
# 돈이 1000원 미만일 경우 걸어간다
else:
    print("걸어간다")

```

```python
# 조건이 몇개인지 먼저 파악하기
#정수 2개 입력받기
#큰수 판별/같을때는 같다고 출력
num1 = int(input())
num2 = int(input())

if num1>num2:
    print("첫번째 정수가 더 큽니다")
elif num2>num1:
    print("두번째 정수가 더 큽니다")
else:
    print("두 수가 똑같습니다")
```

```python
score= int(input("점수 입력"))

if 100<= score >=90:
    print("A학점 입니다")
elif 80<= score> 90:
    print("B학점 입니다")
elif 70<= score >80:
    print("c학점 입니다")
elif 60<= score >70:
    print("D학점 입니다")
else:
    print("F학점 입니다")
```

```python
score= int(input("점수 입력"))

if 100<= score >=90:
    grade = "A"
elif 80<= score:
    grade = "B"
elif 70<= score:
    grade = "C"
elif 60<= score:
    grade = "D"
else:
    grade = "F"
    print(f"{score}점은 {grade}학점 입니다")
```

```python
#1. 넣은돈 입력받기
money = int(input("insert coin>> "))
#2. 메뉴선택 입력받기
menu = input("메뉴선택: ")
change = 0
#3. 어떤 메뉴를 선택하는지에 따라서 잔액 계산 (넣은돈 - 음료금액)
if menu == "1"or menu == "콜라":
    # 우리가 입력받은 돈에서 콜라금액(600)원을 빼주면 됨 ->잔액 계산
    change = money - 600
elif menu == "2" or menu == " 사이다":
    change = money - 800
elif menu == "3" or menu == "물":
    change = money - 1000
else:
    print("메뉴를 잘못입력하셨습니다")
    
ch = o = b = 0    
#4. 만약에 돈이 부족할경우 "잔액이 부족합니다."출력해주기
if change <0 :
    print("잔액이 부족합니다.")
#4. 잔액 거스름돈 계산 (1000원 2개, 500원 1개, 100원 2개)
else:
    ch = change // 1000
    o = change % 1000 // 500
    b = change % 1000 % 500 // 100
    print(f'천원: {ch}개, 오백원{o}개, 백원{b}개')
   
```
