### 조건문

#### if 문
- if 조건식 : 
- (들여쓰기)실행문장
- 주의사항
    - 조건식에는 논리형결과를 써줘야한다! (True,False)
    - 반드시 콜론!!!!
    - 반드시 들여쓰기!!! -> 조건식의 실행범위 영역을 표시
    

```python
# 조건식이 True 였을경우
if True :
    print("실행문장실행")
```

```python
# 조건식이 False 였을때는 아무실행도 하지않고 끝난다
if False: 
    print("실행문장실행")
```

- 들여쓰기의 중요성!

```python
if True:
    print("조건식이 참이었을때 실행문장")
print("if문 바깥쪽에 있는 문장")
```

```python
if False :
    print("조건식이 거짓이었을때 실행문장")
print("if문 바깥쪽에 있는 문장")
# 들여쓰기가 되어있지 않으면 if 의 영향을 받지 않는다
```

```python
money = 1000
if money >= 10000:
    print("택시를탄다")
```

#### if else 문
- if 조건식:
- (들여쓰기)실행문장1
- else :
- (들여쓰기)실행문장2
- 주의사항
    - if, else는 같은선상에 있어야한다!
    - 콜론, 들여쓰기!!
    - else는 참이 아니였을 모든경우를 포함하기때문에 조건식을 적지 않는다!!
    

```python
# 단순if 문과의 차이점: 조건식이 False 였을경우 실행문장의 유무차이!
if False:
    print("실행문장1")
else:
    print("실행문장2")
```

```python
money = 5000
if money >= 10000:
    print("택시를 탄다~")
else:
    print("버스를탄다ㅠㅠ")
```

```python
num = int(input("정수입력>> "))
if num % 3 == 0 and num % 5 ==0:
    print("3과 5의 배수입니다")
else:
    print("3과 5의 배수가 아닙니다~")
```

```python
#1. 상품의 개수 입력받기
count = int(input("사려는 상품의 개수를 입력하세요>> "))
#2 조건문을 사용해서 금액계산
# 11개 미만 -> 10000원
if count <11 :
    price = 10000 * count
# 11개 이상 -> 9000원
else:
    price = 9000 * count
print(f"가격은 {price}원 입니다.")
```

### if elif else 문
- 여러개의 조건을 사용할때

- 주의
    - 반드시 else문을 작성해야한다 XXXX

```python
# 돈이 10000원 이상일경우 택시를 탄다
# 돈이 10000원 미만 1000원 이상일경우 버스를 탄다
# 돈이 1000원 미만일경우 걸어간다
money = 110000
if money >=10000:
    print("택시를 탄다~🤑")
elif 1000<= money <10000 :
    print("버스를탄다ㅠㅠ")
else:
    print("걸어간다😒")
```

```python
# 조건이 몇개인지 먼저 파악하기! (조건이 3개!)
# 1.정수 2개 입력받기
num1 = int(input("첫번째 정수입력"))
num2 = int(input("두번째 정수입력"))
# 2.큰수 판별 / 같을때는 같다고 출력
if num1 > num2 :
    print("첫번째 정수가 더 큽니다!")
elif num2 > num1:
    print("두번째 정수가 더 큽니다.")
else:
    print("두수가 똑같습니다😍")
```

```python
#1. 조건이 몇개인지 확인(5개의 조건 --> elif 문 사용)
#2. score 라는 변수에 점수 입력
score = int(input("점수 입력>> "))
#3. 학점판별하는 조건문 작성
if 90 <= score <= 100:
    result = "A"
elif score >= 80:
    result = "B"
elif score >= 70:
    result = "C"
elif score >=60:
    result = "D"
else:
    result = "F"
print(f"{score}점은 {result}학점입니다.")
```

```python
#1. 넣은돈 입력받기
money = int(input("insert coin>> "))
#2. 메뉴선택 입력받기
menu = input("메뉴선택: ")
change = 0
#3. 어떤 메뉴를 선택하는지에 따라서 잔액 계산 (넣은돈 - 음료금액)
if menu == "1"or menu == "콜라":
    # 우리가 입력받은 돈에서 콜라금액(600)원을 빼주면 됨 ->잔액 계산
    change = money - 600
elif menu == "2" or menu == " 사이다":
    change = money - 800
elif menu == "3" or menu == "물":
    change = money - 1000
else:
    print("메뉴를 잘못입력하셨습니다")
    
ch = o = b = 0    
#4. 만약에 돈이 부족할경우 "잔액이 부족합니다."출력해주기
if change <0 :
    print("잔액이 부족합니다.")
#4. 잔액 거스름돈 계산 (1000원 2개, 500원 1개, 100원 2개)
else:
    ch = change // 1000
    o = change % 1000 // 500
    b = change % 1000 % 500 // 100
    print(f'천원: {ch}개, 오백원{o}개, 백원{b}개')
```
