```python
# 파이썬 라이브러리(모듈)
# 데이터를 분석하는데 용이한 모듈들
#Numpy
```

```python
import numpy as np
```

```python
#리스트,튜플,딕셔너리 = 자료구조
#자료구조들의 특징: 여러가지 값을 저장하는데 목적이 있다.
#numpy내부의 자료구조가 존재: array
#array는 연산에도 목적이 있다
```

```python
# numpy를 사용하려면 자료구조가 array인것들만 사용
# array 만들기
#1. 리스트로 저장할 값 정리하기 
list1 = [1,2,3,4,5]
#2. 리스트를 array로 변환하기 : np.array(list1)
arr1= np.array(list1)
```

```python
list2 = [[1,2,3],[4,5,6]]
arr2 = np.array(list2)
arr2
```

```python
#array(배열) 정보 출력하기
# 배열의 크기
arr2.shape #->(2,3) 2차원은 2개 1차원은 3개가 있다
```

```python
# 배열의 데이터 갯수
arr2.size
```

```python
# 배열의 차원 확인
arr2.ndim
```

```python
list3 = [[1,2],[3,4]],[[5,6],[7,8]]
arr = np.array(list3)
arr
print(f'배열의 크기 :{arr.shape}')
print(f'배열의 차원 :{arr.ndim}')
print(f'배열의 개수 :{arr.size}')
```

```python
#0으로 초기화
#배열의 생성: np.zeros(배열의 크기)
np.zeros((3,4))
```

```python
#1로 초기화: np.ones(배열의 크기)
np.ones((3,4))
```

```python
# 원하는 값으로 배열 생성하기np.full(배열의 크기, 원하는 값)
np.full((3,4),7)
```

```python
# 순서가 있는 값으로 배열 생성하기
#1~50까지가 담긴 배열 생성하기 : np.arange(시작숫자,끝숫자+1,간격)

np.arange(1,51,1)
```

```python
# 랜덤값으로 만들기
#0~1까지 실수값으로 만들기: np.random.rand(배열의 크기)
np.random.rand(2,3)
```

```python
#랜덤값으로 만들기
#원하는 범위의 숫자 안에서 정수의 값으로 랜덤으로 만들기
#np.random.randint(범위의 첫 숫자,범위의 마지막 숫자+1,size =배열의크기)
np.random.randint(2,10,size = (2,3))
```

```python
#0~9까지 랜덤한 정수 만들기
#0~1까지 랜덤한 실수 만들기
#둘이 더하기

result1 = np.random.randint(0,10,size =(3,2))
result2 = np.random.rand(3,2)
result1+result2
```

```python
 #0~1까지의 랜덤한 실수값*10
np.random.rand(3,2)*10
```

```python
#배열에서 원하는 값 가져오기
#0~49까지의 값이 담긴 배열
#(5.10)의 크기를 갖도록 만들어주자: reshape(변경 될 크기)
# reshape 전과 후의 데이터 갯수가 동일해야함
arr3 = np.arange(0,50,1).reshape((5,10))
arr3
```

```python
#배열에서도 인덱싱과 슬라이싱이 가능하다
#배열도 순서가 있는 자료의 구조
arr3[22:30]
```

```python
#2차원 배열에서 원하는 값 출력하기
#인덱싱, 슬라이싱 사용, 차원 별로 생각
arr3
```

```python
# arr3[3차원 인덱스값, 1차원 인덱스값]
arr3[1,2:5]
```

```python
arr3[0:5,0]
```

```python
arr3[1:4,2:5]
```

```python
arr3[:-1,:5]
```
