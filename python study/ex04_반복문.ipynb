```python
import random
```

```python
randomNum = random.randint(1,50)
print(randomNum)

while True :
    inputNum = int(input("숫자를 입력하세요 >> "))
    if randomNum > inputNum :
        print(f'{inputNum}보다 큰 수 입니다.')
    elif randomNum < inputNum :
        print(f'{inputNum}보다 작은 수 입니다.')
    else :
        print('정답을 맞추셨습니다.')
        break
```

```python
randomNum2 = random.randint(1,50)
print(randomNum2)

inputNum2 = randomNum2-1

while randomNum2 != inputNum2 :
    inputNum2 = int(input('숫자를 입력하세요 >> '))
    if randomNum2 > inputNum2 :
        print(f'{inputNum2}보다 큰 수 입니다.')
    elif randomNum2 < inputNum2 :
        print(f'{inputNum2}보다 작은 수 입니다.')
print('정답을 맞추셨습니다.')
```

### 실습문제


```python
cw = int(input('현재 몸무게 : '))
gw = int(input('목표 몸무게 : '))

week = 1
while cw > gw :
    lw = int(input(f'{week}주차 감량 몸무게 : '))
    cw = cw - lw
    week += 1 # week = week+1
print(f'{cw}kg 달성!! 축하합니다.')
```

```python
# 긍정리뷰/부정리뷰 개수

nr = 0
pr = 0

for i in range(0, len(movie_review), 1) :
    if movie_review[i][1] == 1 :
        pr += 1
    else :
        nr += 1
print(f"긍정 '{pr}'개")
print(f"부정 '{nr}'개")
```

```python
print("="*40)
studentScore = []
while True : 
    select = int(input('[1]시험점수입력 [2]성적확인 [3]종료 >> '))
    if select == 3 :
        print("="*40)
        break
    elif select == 1 :
        name = input('이름입력 : ')
        javaScore = int(input('Java 점수 입력 : '))
        webScore = int(input('Web 점수 입력 : '))
        studentScore.append([name, javaScore, webScore])
        
    elif select == 2 :
        print('이름\tJava\tWeb\tGrade')
        for li in studentScore :
            avg = (li[1]+li[2])/2 
            if avg >= 90 :
                grade = "A"
            elif avg >= 80 :
                grade = "B"
            elif avg >= 70 :
                grade = "C"
            elif avg >= 60 :
                grade = "D"
            else :
                grade = "F"
            print(f'{li[0]}\t{li[1]}\t{li[2]}\t{grade}')
    print("-"*40)
```
