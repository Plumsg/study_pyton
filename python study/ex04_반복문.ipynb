```python
import random
random_num = random.randint(1,51)
print(random_num)

while True:
    num =int(input("숫자를 입력하세요>>"))

    if random_num == num:
        print("정답을 맞추셨습니다")
        break
    elif random_num > num:
        print(f"{num}보다 큰 수 입니다")
    else :
        print(f"{num}보다 작은 수 입니다")
       
```

```python
week = 0
num =int(input("현재 몸무게"))
num2=int(input("목표 몸무게"))

while True:
    week+=1
    wkg = int(input())
    print(f"{week}주차 감량 몸무게 : {wkg}")
    num -= wkg       
    if num <= num2:
          break
print(f"{num}kg 달성!! 축하합니다")
        
```

```python
cw=int(input("현재 몸무게 : "))
gw=int(input("목표 몸무게 : "))
week = 1
while cw >gw :
    lw = int(input(f'{week}주차 감량 몸무게 : '))
    cw = cw-lw 
    week+= 1 # week는 week+!
print(f'{cw}kg 달성!! 축하합니다')
    
```

```python
score_list = [90,45,70,60,55]
cnt= 1
for a in score_list:
    
    if a >=60:
        print(f"{cnt}번 학생은 합격입니다")
    else:
        print(f"{cnt}번 학생은 불합격입니다")
    cnt+=1
```

```python
startNum = int(input("Start: "))
endNum = int(input('End: '))
for i in range(startNum,endNum+1):
    num = 0
    print(f'{i}의 약수 :', end = "")
    for j in range(1,i+1):
        if i % j ==0:
            num += j
    print(num)
```
