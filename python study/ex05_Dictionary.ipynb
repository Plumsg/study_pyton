

immutable : 불변의.

key값은 유일해야 한다. (동일한 값을 가지면 안됨)

```python
# 딕셔너리명 = {key:value, key2:value2...} // 튜플(), 리스트[]
# key는 변하지 않는 값을 사용
# value는 변하는 값, 변하지 않는 값 모두 사용 가능
# value값에는 문자형, 숫자형, 튜플, 리스트 형태 상관 없이 모두 가능
```

```python
a = {}
b = {"name":"KinYujin", "pn":"010-1111-2222", "age":20}
c = {1:"asd", 2:2135}
print(type(b))
print(b)
print(c)
```

```python
# 딕셔너리 값 추가하기
# 딕셔너리명[key] = value
dic1 = {"name":"KimYujin", "pn":"010-1111-2222", "age":20}
dic1['birth'] = '0531'
dic1['pn'] = "010-3333-4444"
print(dic1)
print(dic1["name"])
```

```python
# 딕셔너리 값 가져오기 1
print(dic1["name"])
# print(dic1["email"]) -> KeyError

# 딕셔너리 값 가져오기 2
print(dic1.get("name"))
print(dic1.get("email"))

#get 함수의 사용
if dic1.get('email') == None :
    print("없는 데이터")
```

```python
# 딕셔너리 key만 가져오기
# 딕셔너리명.keys()

keyLi = dic1.keys() 
# dict_keys(['name', 'pn', 'age', 'birth'])

keyLi = list(dic1.keys())
#['name', 'pn', 'age', 'birth']
```

```python
for i in dic1.keys() :
    print(i)
```

```python
# 딕셔너리 value만 가져오기
# 딕셔너리명.value()

valueLi = list(dic1.values())
print(valueLi[0])

for value in dic1.values() :
    print(value)
```

```python
dic1.items()
```

```python
for key,value in dic1.items() :
    print(key, value)
```

```python
# 딕셔너리와 in 키워드 사용
print('name' in dic1)
print('email' in dic1)
print('20' in dic1)
# => value로는 in 키워드로 찾을 수 없다. False나옴.
```

```python
# 딕셔너리 값 삭제하기
# del 딕셔너리명[key]
del dic1['age']
```

```python
print(dic1)
```

```python
#딕셔너리 모두 지우기
dic1.clear()
```
