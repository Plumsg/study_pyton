```python
#리뷰 길이 평균 값
#  sum_movie = sum(movie_review)/len(movie_review)
result = 0
for i in movie_review:
    result +=len(i[0])
print(result/len(movie_review))
```

```python
#긍정/부정 리뷰 갯수
```

```python
cnt1 = 0
cnt0 = 0
for i in movie_review:
    if i[1] ==0:
        cnt0 +=1
    else:
        cnt1 +=1
print(cnt1,cnt0)

print(f"긍정 '{cnt1}'개")
print(f"부정 '{cnt0}'개")
print("똑바로 해라 김유진바보")
    
```

```python
# 리뷰 길이 최대 값/내용

max_list= len(movie_review[0][0])


for i in movie_review:
    if max_list<len(i[0]):
        max_list = len(i[0])
        max_value = i[0]
    
print(max_list)
print(max_value)
```
