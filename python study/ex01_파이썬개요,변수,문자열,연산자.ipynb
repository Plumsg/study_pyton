### 문자열 인덱싱과 슬라이싱
 - 문자열은 처음부터 끝까지 인덱스를 배정 받는다
     - 주의1. 인덱스는 0번부터 시작한다
     - 주의2. 띄어쓰기도 인덱스를 가진다
 - 인덱싱: 문자열 안에서 내가 원하는 데이터 하나를 뽑아올때 사용!
     - 인덱싱 방법: 변수명[인덱스번호]
 - 슬라이싱: 문자열 안에서 범위를 주고 잘라올때 사용!
     - 슬라이싱 방법: 변수명.[시작인덱스:끝인덱스+1]
     - 주의 내가 가져오고 싶은 범위에서 끝인덱스+1 해주어야한다

```python
name = "My name is SG"
# 인덱싱 방법: 변수명[인덱스 번호]
# n 가져오기name = "My name is SG"
print(name[0:2])

print(name[3])
# 대문자 S (이름앞글자) 가져오기
print(name[-2])
```

```python
# 슬라이싱
#슬라이싱 방법: 변수명[시작인덱스:끝인덱스+1]
name = "My name is SG"
print(name[0:2])
```

```python
name = "My name is SG"
print(name[0:2])
print(name[3:7])
print(name[-2:])

```

```python
# name is SG 가져오기
print(name[3:])
```

### 포매팅
- 문자열 안에 특정한 값(변수)을 바꿔야 할 때 사용한다
- 3가지 방법
    - %기호 포매팅 -> 단점: 자료형에 따라서 기호가 달라진다.
    - format 함수 
    - f 문자열 포매팅 -> 장점: 직관적이다
 

```python
s5 = "오늘은 6월 9일 입니다!"
s5
```

```python
# % 기호 포매팅방법
# 단점: 포맷코드를 암기해야 사용가능
month = 6
day = 9
s6 = "오늘은 %d월 %d일 입니다!"%(month,day)
print(s6)
```

```python
# format 함수를 활용한 포매팅
month = 6
day = 9
s7 = "오늘은 {}월 {}일 입니다".format(month,day)
print(s7)
```

```python
# f 문자열포매팅   
month = 6
day = 9
s8 = f"오늘은 {month}월 {day}일 입니다!"
print(s8)
```

```python
x = 100
y = 200
add = x+y
#%d기호 포매팅
print("%d와 %d의 합은 %d입니다"%(x,y,add))
#format 함수를 사용한 포매팅
print("{}와 {}의 합은 {}입니다".format(x,y,add))
# f문자열 포매팅
print(f"{x}와 {y}의 합은 {x+y}입니다")
```

### 논리연산자
- 논리형(True,False) 결과값을 연산해주는 연산자
    - and : 두개의 값 '모두'가 True 였을때 True 라는 결과가 출력
    - or : 둘중 '하나라도' True 라면 True 라는 결과를 출력
    - not: 논리를 부정하는 연산자 (논리 앞에 써주기)

```python
a= 10
b = 7
print(a!=b and a==b)
print(a!=b or a==b)
print(a>b)
print(not a>b)

```

### 삼향연산자
- 조건식의 결과가 참이라면 a를 실행
- 조건식의 결과가 거짓이라면 b를 실행
- 주의
    - 조건식의 결과는 항상 논리형(True,False)

```python
# 나이가 20살 이상이면 "성인"을 20살 미만이면 "미성년자"를 출력

age = 18

"성인" if age >=20 else "미성년자"
```
