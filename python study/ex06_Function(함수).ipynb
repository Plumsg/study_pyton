```python
def divisor(x) :
    result = []
#     x = int(x) <-   ~!
    for i in range(1,x+1,1) :
        if x%i == 0:
            result.append(i)
    return result
```

```python
def close10(n1,n2) :
    if abs(n1-10)>abs(n2-10) :
        return n2
    else :
        return n1
```

```python
num1 = int(input("    : "))
num2 = int(input("    : "))
result = close10(num1,num2)

print("10   : ",result)
```

```python
def closeTen(n1,n2) :
    ch1 = n1-10
    ch2 = n2-10
    if ch1*ch1 > ch2*ch2 :
        return n2
    else :
        return n1
```

```python
closeTen(-3,-4)
```

```python
#   ,   F

def isDuplicate(li):
    for i in range(0,len(li)-1,1):
        for k in range(i+1,len(li),1):
            if li[i] == li [k]:
                return True
    return False
```

```python
li1 = [1,2,3,4,5,6,7,8]

print(isDuplicate(li1))
```

```python
li3 = [5,5,6,7,8,8]

print(isDuplicate(li3))
```

```python
import random

def createRandomNumber(num) :
    while True :
        totalLi = []
        for i in range(0,num):
            totalLi.append(random.randint(1,5))
        if not isDuplicate(totalLi):
            return totalLi
```

```python
print(createRandomNumber(3))
```

```python
import random

def createRandomNumber(cnt) :
    li = [0,0,0] # True   for 
    while isDuplicate(li):
        li = []
        for i in range(cnt) :
            more = random.randint(1,5)
            li.append(more)
    return li
```

```python
creatRandomNumber(2)
```

```python
def strikeBallCounting(li1,li2):
    strike = 0
    ball = 0
    
    for i in range(0,len(li1),1):
        for k in range(0,len(li2),1):
            if li1[i]==li2[k]:
                if i == k :
                    strike += 1
                else :
                    ball += 1
                    
    return strike, ball
```

```python
print(strikeBallCounting([1,2,3],[1,5,6]))
```

```python
print("==GameStart==")
n = int(input("?"))
randomList = createRandomNumber(n)
print(randomList)

while True :

    inputList = []
    for i in range(0,len(randomList),1) :
        inputList.append(int(input(f"{i+1}   : ")))

    strike,ball = strikeBallCounting(randomList,inputList)
    print(f"strike : {strike}\t ball:{ball}")
    
    if strike == n :
        break
        
    print("")
```
