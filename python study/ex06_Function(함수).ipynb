```python
def divisor(x) :
    result = []
#     x = int(x) <- 없어도 되는 문장이에요~!
    for i in range(1,x+1,1) :
        if x%i == 0:
            result.append(i)
    return result
```

```python
divisor(15)
```

```python
def cal(num1,num2,op):
    """ 덧셈과 뺄셈을 계산하는 함수   """
    if op == "+" :
        total = num1 + num2
    elif op == "-" :
        total = num1 - num2
        
    return total
```

```python
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")
result = cal(num1,num2,op)
print(f"결과 : {result}")  
#print("결과 : ",cal(num1,num2,op))

# 연산자 +,- 가 아닌 다른 값을 넣으면? => None
```

```python
def close10(n1,n2) :
    if abs(n1-10)>abs(n2-10) :
        return n2
    else :
        return n1
```

```python
num1 = int(input("첫 번째 정수 입력 : "))
num2 = int(input("두 번째 정수 입력 : "))
result = close10(num1,num2)

print("10에 가까운 수 : ",result)
```

```python
def closeTen(n1,n2) :
    ch1 = n1-10
    ch2 = n2-10
    if ch1*ch1 > ch2*ch2 :
        return n2
    else :
        return n1
```

```python
closeTen(-3,-4)
```

```python
# 가변 매개변수를 활용해 모든 숫자를 더해서 반환하는 함수를 정의
def add(*args) :
    total = 0
    for num in args : 
        total += num
    return total
```

```python
print(add(1,2,3))
print(add(1,2,3,4,5,6,7,8,9))
```

```python
# 일반 매개변수와 가변매개변수를 합쳐서 사용하고 싶으면 가변매개변수 앞에 사용하자!!
def add(n,*args) :
    print(n)
    total = 0
    for num in args : 
        total += num
    return total
```

```python
print(add(1,2,3))
print(add(1,2,3,4,5,6,7,8,9))
```

```python
#중복되는게 하나라도 있으면 트루, 하나도 없으면 F

def isDuplicate(li):
    for i in range(0,len(li)-1,1):
        for k in range(i+1,len(li),1):
            if li[i] == li [k]:
                return True
    return False
```

```python
li1 = [1,2,3,4,5,6,7,8]

print(isDuplicate(li1))
```

```python
li3 = [5,5,6,7,8,8]

print(isDuplicate(li3))
```

```python
import random

def createRandomNumber(num) :
    while True :
        totalLi = []
        for i in range(0,num):
            totalLi.append(random.randint(1,5))
        if not isDuplicate(totalLi):
            return totalLi
```

```python
print(createRandomNumber(3))
```

```python
import random

def createRandomNumber(cnt) :
    li = [0,0,0] #중복으로 True를 반환해야 밑의 for문이 시작하니까
    while isDuplicate(li):
        li = []
        for i in range(cnt) :
            more = random.randint(1,5)
            li.append(more)
    return li
```

```python
creatRandomNumber(2)
```

```python
def strikeBallCounting(li1,li2):
    strike = 0
    ball = 0
    
    for i in range(0,len(li1),1):
        for k in range(0,len(li2),1):
            if li1[i]==li2[k]:
                if i == k :
                    strike += 1
                else :
                    ball += 1
                    
    return strike, ball
```

```python
print(strikeBallCounting([1,2,3],[1,5,6]))
```

```python
print("==GameStart==")
n = int(input("난이도는?"))
randomList = createRandomNumber(n)
print(randomList)

while True :

    inputList = []
    for i in range(0,len(randomList),1) :
        inputList.append(int(input(f"{i+1}번째 숫자 입력 : ")))

    strike,ball = strikeBallCounting(randomList,inputList)
    print(f"strike : {strike}\t ball:{ball}")
    
    if strike == n :
        break
        
    print("")
```
