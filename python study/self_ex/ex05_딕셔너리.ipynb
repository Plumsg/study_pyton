```python
def divisor(num):
    num_list = []
    for i in range(1,num+1):
        if num%i ==0:
            num_list.append(i)
    return num_list
```

```python
def cal(num1,num2,op):
    if op == "+":
        return num1+num2
    elif op == "-":
        return num1-num2

num1 = int(input("첫 번째 정수 입력>>"))
num2 = int(input("두 번째 정수 입력>>"))
op= (input("연산자 입력(+,-)>>"))

result = cal(num1,num2,op)
print("결과:{}".format(result))

```

```python
# 가변 매개변수를 활용해 모든 숫자를 더해서 반환하는 함수를 정의
def add(*args):
    total = 0
    for num in args:
        total+= num
    return total# 튜플 타입
```

```python
print(add(1,2,3))
```

```python
def isDuplicate(li):
    for k in range(0,len(li)-1,1):
            for i in range(k+1, len(li),1):
                if li[k] == li[i]:
                    return True
    return False
```

```python
import random
```

```python
def createRandomNumber (cnt):
    randomList = [0,0,0]
    while isDuplicate(randomList):
        randomList = []
        for i in range(cnt):
            num = random.randint(1,5)
            randomList.append(num)
    return randomList
        
```

```python
createRandomNumber(3)
```

```python
def strikeBallCounting(li1,li2):
    strike = 0
    ball = 0
    \
    for k in range(len(li1)):
        for i in range(len(li2)):
            if li1[0] == li2[i]:
                if k == i :
                    strike += 1
                else:
                    ball+= 1
    return strike, ball
            
```

```python
strikeBallCounting([1,2,3],[3,2,1])
```

```python
print("==GameStart")
randomList =createRandomNumber(3)
print(randomList)

while True:

    inputList = []
    for i in range(len(randomList)):
        num = int(input(f'{i+1}번째 숫자 입력 :'))
        inputList.append(num)
    strike, ball = strikeBallCounting(randomList, inputList)
    print(f"strike: {strike}\tball : {ball}")
    if strike == len(randomList):
        print("게임종료!!")
        break
```
